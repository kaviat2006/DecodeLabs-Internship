```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("Dataset for Data Analytics (3).xlsx")
df.head()
```

```python
df.groupby("Product")["TotalPrice"].sum().plot(kind="bar")
plt.title("Total Sales by Product")
plt.savefig("bar_chart.png")
plt.show()
```

```python
df["Date"]=pd.to_datetime(df["Date"])

df.groupby("Date")["TotalPrice"].sum().plot(kind="line")
plt.title("Sales Trend")
plt.savefig("line_chart.png")
plt.show()
```

```python
df["PaymentMethod"].value_counts().plot(kind="pie",autopct="%1.1f%%")
plt.ylabel("")
plt.title("Payment Methods")
plt.savefig("pie_chart.png")
plt.show()
```

```python
plt.hist(df["TotalPrice"],bins=20)
plt.title("Distribution of Total Price")
plt.xlabel("Total Price")
plt.ylabel("Frequency")
plt.savefig("histogram.png")
plt.show()
```

```python
plt.scatter(df["Quantity"],df["TotalPrice"])
plt.xlabel("Quantity")
plt.ylabel("Total Price")
plt.title("Quantity vs Total Price")
plt.savefig("scatter_plot.png")
plt.show()
```
