```python
df=pd.read_excel("Dataset for Data Analytics (2).xlsx")
df.head()
```

```python
sqldf("SELECT * FROM df LIMIT 10")
```

```python
sqldf("SELECT COUNT(*) AS Total_Records FROM df")
```

```python
sqldf("""
SELECT Product, SUM(Quantity)AS Total_Quantity
FROM df
GROUP BY Product
ORDER BY Total_Quantity DESC
""")
```

```python
sqldf("""
SELECT *
FROM df
WHERE OrderStatus='Delivered'
""")
```

```python
sqldf("""
SELECT OrderID,
Product,
TotalPrice
FROM df
ORDER BY TotalPrice DESC
LIMIT 10
""")
```

```python
sqldf("""
SELECT PaymentMethod,
SUM(TotalPrice) AS Sales
FROM df
GROUP BY PaymentMethod
ORDER BY Sales DESC
""")
```

```python
sqldf("""
SELECT Product,
AVG(Quantity) AS Average_Quantity
FROM df
GROUP BY Product
ORDER BY Average_Quantity DESC
""")
```
